```python
#1 Introduction: Formulate a research question that allows for explorations and comparisons.
#The research question:
#Has The Netherlands compared to other Western Europe countrys less deaths in 2021? And is there a corralation with vaccinated people?
```

```python
#2 Data preparation: The dataset of COVID
# pandas module for working with Series and Dataframes
import pandas as pd

#importing altair and maxrows
import altair as alt

#Max rows
alt.data_transformers.disable_max_rows()

# load dataset
covid = pd.read_csv("https://covid.ourworldindata.org/data/owid-covid-data.csv")

covid.head()
```

```python
#3 Explore and clean the data by: (20 pt)
# The Netherlands total deaths with other Western Countrys and there total vaccinated peoples.

#First Select on several Locations (= Countrys)
country_small_set = ['NLD', 'GBR', 'BEL', 'LUX', 'DEU', 'DNK', 'FRA']

#Print the selction
small_df = covid.loc[covid.iso_code.isin(country_small_set)]


small_df
```

```python
#New dataset 

#Secondly select the columns and make final dataframe
final_df = small_df.loc[:, ['iso_code', 'total_deaths', 'total_vaccinations', 'date']]

final_df
```

```python
#Getting grid of the NaN in my final_dataset

final_df.dropna(subset=['total_deaths'],inplace=True)
final_df.dropna(subset=['total_vaccinations'],inplace=True)
final_df.dropna(subset=['date'],inplace=True)

final_df
```

```python
#Switching datatype from Object to Datetime with 'Date'

final_df['date'] = pd.to_datetime(final_df['date'])

final_df.dtypes
```

```python
#4 Describe and visualise: (50 pt)

alt.Chart(final_df).mark_line().encode(
     x='date',
     y='total_deaths',
      color='iso_code')

#Here is the total deaths from 2020 December until 2021 showed per country
#Conclusion is: The Netherlands, compared to other western countrys, has a low death count
#Lets compare this graph with the total vaccinations if there is a connection between these two.

#import widgets and vega_datasets
import ipywidgets as widgets
from vega_datasets import data

source = data.stocks()
```

```python
#Lets compare the total date with the total vaccinations per country

alt.Chart(final_df).mark_line().encode(
    x=alt.X('date', title='Date'),
    y=alt.Y('total_vaccinations', title='Total Vaccinations'),
    color='iso_code',
    strokeDash='iso_code',
    tooltip=['date', 'total_vaccinations'] 
    )


```

```python
#Lets compare the total death with total vaccinated people per country

alt.Chart(final_df).mark_line().encode(
    x=alt.X('date', title='Date'),
    y=alt.Y('total_deaths', title='Total Deats'),
    color='iso_code',
    tooltip=['date', 'total_deaths'] )
```

```python


alt.Chart(final_df).mark_line().encode(
    x=alt.X('total_vaccinations', title='Total Vaccinations'),
    y=alt.Y('total_deaths', title='Total Deaths'),
    color='iso_code',
    strokeDash='iso_code',
    tooltip=['total_deaths', 'total_vaccinations'] )
```
