```python
#1 Introduction: Formulate a research question that allows for explorations and comparisons.
#The research question:
#Has The Netherlands, compared to other Western Europe countrys, less deaths in 2021 in relation to the total vaccinations? 
```

```python
#2 Data preparation: The dataset of COVID
# pandas module for working with Series and Dataframes
import pandas as pd

#importing altair and plotly
import altair as alt
import numpy as np

#Max rows
alt.data_transformers.disable_max_rows()

# load dataset
covid = pd.read_csv("https://covid.ourworldindata.org/data/owid-covid-data.csv")

covid.head()
```

```python
#3 Explore and clean the data by: (20 pt)
# The Netherlands total deaths with other Western Countrys and there total vaccinated peoples.

#First Select on several Locations (= Countrys)
country_small_set = ['NLD', 'GBR', 'BEL', 'LUX', 'DEU', 'DNK', 'FRA']

#Print the selction
small_df = covid.loc[covid.iso_code.isin(country_small_set)]


small_df
```

```python
#New dataset 

#Secondly select the columns and make final dataframe
final_df = small_df.loc[:, ['iso_code', 'total_deaths', 'total_vaccinations', 'date']]

final_df
```

```python
#Getting grid of the NaN in my final_dataset

final_df.dropna(subset=['total_deaths'],inplace=True)
final_df.dropna(subset=['total_vaccinations'],inplace=True)
final_df.dropna(subset=['date'],inplace=True)

final_df
```

```python
#Switching datatype from Object to Datetime with 'Date'

final_df['date'] = pd.to_datetime(final_df['date'])

final_df.dtypes
```

```python
#4 Describe and visualise: (50 pt)

#import widgets and vega_datasets
import ipywidgets as widgets
```

```python
final_df.describe()
```

```python
final_df.groupby(['iso_code'])['total_deaths', 'total_vaccinations'].mean()
```

```python
#In this chart it is shown that: The Netherlands has less then other compared countrys.
```

```python
#Lets compare the total date with the total vaccinations per country

MYchart_1 = alt.Chart(final_df).mark_line().encode(
    x=alt.X('date', title='Date'),
    y=alt.Y('total_vaccinations', title='Total Vaccinations'),
    color='iso_code',
    strokeDash='iso_code',
    tooltip=[
            alt.Tooltip('date', title= 'Date'),
            alt.Tooltip('total_vaccinations', title= 'Total Vaccinations')]
)
display(MYchart_1)

```

```python
#Iso_code abbreviation
#BEL = Belgium, DEU = Germany, DNK = Denmark, FRA = France, GBR = Great Brittain, LUX = Luxenburg, NLD = The Netherlands
```

```python
# In the chart MYchart_1 we see the total vaccinations per country from decemeber until the beginning of 2022. Germany, Great Brittain and France are leading with total vaccinations
# We must keep in mind that the population size matters in this case. Belgium, Luxenburg, Denmark are smaller in population then the other western countrys.
# The Netherlands has a fluent rising vaccinated population. In total 24 million vaccinations are set are given.
```

```python
#Lets compare the total death with total vaccinated people per country
```

```python
MYchart_2 =  alt.Chart(final_df).mark_line().encode(
    x=alt.X('date', title='Date'),
    y=alt.Y('total_deaths', title='Total Deats'),
    color='iso_code',
    tooltip=[
            alt.Tooltip('date', title= 'Date'),
            alt.Tooltip('total_deaths', title= 'Total Deaths')]
    )

display(MYchart_2)
```

```python
# MYchart_1 is telling us wich countrys are having the most deaths. 
# The leading country, with total deaths, is Great Brittain. As second we have France and as third Germany.
# Taking in mind that Great Brittain has 68 million citizens and Germany 83 million citizens. We can conculude that Great brittian has more deaths compared to the population of these countrys.
# Next we will look at The Netherlands. We can conclude that there is a small rising in total deats over the period of one year. Other countrys have big rising deaths but have a bigger population.
```

```python


MYchart_3 =  alt.Chart(final_df).mark_line().encode(
    y=alt.Y('total_vaccinations', title='Total Vaccinations'),
    x=alt.X('total_deaths', title='Total Deaths'),
    color='iso_code',
    strokeDash='iso_code',
    tooltip=[
            alt.Tooltip('total_vaccinations', title= 'Total Vaccinations'),
            alt.Tooltip('total_deaths', title= 'Total Deaths')]
)

display(MYchart_3)
```

```python
# MYchart_3 is giving us the following information.
# First we can conclude that: When countrys are starting with vaccinations the total deaths are getting less or getting delayed. 
# The Netherlands has the same trend as we already concluded, when starting with vaccinations there ar less deaths then we aren't.
```
